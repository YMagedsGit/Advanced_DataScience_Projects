1. __Import__ the json files __blockbusters.json__, __blockbusters2.json__, __blockbusters3.json__ and load the datasets into Pandas DataFrames.


```python
blockbuster = pd.read_json('blockbusters.json')
blockbuster1 = pd.read_json('blockbusters2.json')
movies = pd.read_json('movies.json')
```

```python
collection_df = pd.json_normalize(movies['belongs_to_collection'])

# Rename columns to avoid collisions (optional but good practice)
collection_df.columns = [f'collection_{col}' for col in collection_df.columns]

# Combine back with the original dataframe
movies = pd.concat([movies, collection_df], axis=1)

#genres
movies['genres'] = movies['genres'].apply(lambda x: [d['name'] for d in x] if isinstance(x, list) else [])
```
