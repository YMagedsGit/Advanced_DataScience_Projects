## Importing & merging many files (easy case)

3. __Load__ all files (yob????.txt) and __merge/concatenate__ all files into one Pandas DataFrame. Make sure you add the __column "Year"__.

```python
files = glob.glob('yob*.txt')
df_list = []

for f in files:
    # Extract the year from the filename (e.g., movies_2018.csv → 2018)
    filename = os.path.basename(f)
    year = filename.replace('yob', '').replace('.txt', '')

    # Read file
    temp_df = pd.read_csv(f, header = None, names = ["Name", "Gender", "Count"])

    # Add year column
    temp_df['year'] = int(year)

    # Append to list
    df_list.append(temp_df)

# Combine all years into one DataFrame
df = pd.concat(df_list, ignore_index=False)
```

4. Create a __RangeIndex__ and __save__ the DataFrame (with columns "Year", "Name", "Gender" and "Count") in a new csv-file. 

```python
df.to_csv('my_us_baby_names.csv',index=False)
```
